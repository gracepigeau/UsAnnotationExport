```python
output_image_size = 512

train_folder = r"scoliosis\train\combine"
test_folder = r"scoliosis\test\combine"

model_output_folder = r"output\scoliosis_model"
image_output_folder = r"output\scoliosis_test"

import tensorflow as tf
```

```python
import os

from local_vars import root_folder

train_fullpath = os.path.join(root_folder, train_folder)
test_fullpath = os.path.join(root_folder, test_folder)

output_fullpath = os.path.join(root_folder, image_output_folder)

image_file_list_train = [f for f in os.listdir(train_fullpath) if f.endswith('.png')]
image_file_list_test = [f for f in os.listdir(test_fullpath) if f.endswith('.png')]


num_images_train = len(image_file_list_train)
num_images_test = len(image_file_list_test)

print( "Found {} training image files".format(num_images_train))
print( "Found {} test image files".format(num_images_test))
```

```python
import cv2
import numpy as np

from ipywidgets import IntProgress
from IPython.display import display

input_intensity_scaling = 1/255.0

images_folder =  os.path.join(image_output_folder, "images")

ultrasound_file_list_target = [f for f in os.listdir(images_folder) if f.endswith('-targets.png')]
ultrasound_file_list_generated = [f for f in os.listdir(images_folder ) if f.endswith('-outputs.png')]
segmentation_file_list = [f for f in os.listdir(images_folder) if f.endswith('-inputs.png')]

ultrasound_images_target = np.zeros([num_images_test, output_image_size, output_image_size, 1]).astype(np.float16)
ultrasound_images_generated = np.zeros([num_images_test, output_image_size, output_image_size, 1]).astype(np.float16)
segmentation_images = np.zeros([num_images_test, output_image_size, output_image_size, 1]).astype(np.uint8)

f = IntProgress(min=0, max=num_images_test)
print("Reading files and saving them as numpy arrays...")
display(f)

for i in range(num_images_test):
    f.value = i
    current_us_target_file_name = ultrasound_file_list_target[i]
    current_us_target_file_fullname = os.path.join(images_folder, current_us_target_file_name)
    
    current_us_generated_file_name = ultrasound_file_list_generated[i]
    current_us_generated_file_fullname = os.path.join(images_folder, current_us_generated_file_name)
    
    current_segmentation_file_name = segmentation_file_list[i]
    current_segmentation_file_fullname = os.path.join(images_folder, current_segmentation_file_name)
    
    ultrasound_target_image = cv2.imread(current_us_target_file_fullname, 0)
    ultrasound_generated_image = cv2.imread(current_us_generated_file_fullname, 0)
    segmentation = cv2.imread(current_segmentation_file_fullname, 0)

    if ultrasound_target_image is None:
        print("File not read: {}".format(current_us_target_file_fullname))
        break
    if ultrasound_generated_image is None:
        print("File not read: {}".format(current_generated_target_file_fullname))
        break
    if segmentation is None:
        print("File not read: {}".format(current_segmentation_file_fullname))
        break
    if ultrasound_target_image.shape[0] != output_image_size or ultrasound_target_image.shape[1] != output_image_size:
        ultrasound_images_target[i,...,0] = cv2.resize(ultrasound_target_image, (output_image_size, output_image_size)).astype(np.float16) * input_intensity_scaling
        ultrasound_images_generated[i,...,0] = cv2.resize(ultrasound_generated_image, (output_image_size, output_image_size)).astype(np.float16) * input_intensity_scaling
        segmentation_images[i,...,0] = cv2.resize(segmentation, (output_image_size, output_image_size))
    else:
        ultrasound_images_target[i,...,0] = ultrasound_target_image.astype(np.float16) * input_intensity_scaling
        ultrasound_images_generated[i,...,0] = ultrasound_generated_image.astype(np.float16) * input_intensity_scaling
        segmentation_images[i,...,0] = segmentation[...]

print("Images converted to numpy arrays")
```

```python
#view test results

import matplotlib.pyplot as plt
from random import sample

np.set_printoptions(threshold=np.inf)

num_test = ultrasound_images_target.shape[0]
num_show = 5

indices = [i for i in range(num_test)]
sample_indices = sample(indices, num_show)

asdf = image_file_list_train[0]

fig = plt.figure(figsize=(18, num_show*5))
for i in range(num_show):
    
    target = ultrasound_images_target[sample_indices[i], :, :, 0]
    prediction = ultrasound_images_generated[sample_indices[i], :, :, 0]
    
    #print(target)
    
    
    a0 = fig.add_subplot(num_show,3,i*3+1)
    img0 = a0.imshow(segmentation_images[sample_indices[i], :, :, 0].astype(np.float32))
    a0.set_title("Segmentation #{} - {}".format(sample_indices[i], ultrasound_file_list_target[sample_indices[i]]))
    
    a1 = fig.add_subplot(num_show,3,i*3+2)
    img1 = a1.imshow(target.astype(np.float32))
    a1.set_title("Ultrasound #{}".format(sample_indices[i]))

    a2 = fig.add_subplot(num_show,3,i*3+3)
    img2 = a2.imshow(prediction.astype(np.float32))
    
    accuracy = np.sum(prediction == target) / (output_image_size*output_image_size) * 100
    
    a2.set_title("Prediction #{} - Accuracy = {}%".format(sample_indices[i], round(accuracy,3)))
    c = fig.colorbar(img2, fraction=0.046, pad=0.04)
    
    
```
